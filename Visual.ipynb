```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x_data = np.random.random(50) * 100
y_data = np.random.random(50)* 100

plt.bar(x_data,y_data)
plt.show()
```

```python
x_data = np.random.random(50) * 100
y_data = np.random.random(50)* 100

plt.plot(x_data,y_data)
plt.show()
```

```python
x_data = np.random.random(50) * 100
y_data = np.random.random(50)* 100

plt.scatter(x_data,y_data,c="#000", marker="*")
plt.show()
```
